```python
def male_female_child(passenger):
    if passenger['Age'] < 16:
        return 'child'
    else:
        return passenger['Sex']
```

```python
titanic_df['Person'] = titanic_df.apply(male_female_child,axis=1)
```

```python
# Chart: Shows the count of passengers by class (Pclass), split by Person (man, woman, child).
sns.catplot(x='Pclass',data=titanic_df,hue='Person',kind ='count')
```

```python
plt.show()
```

```python
# Chart: KDE plot showing age distribution by sex.
fig = sns.FacetGrid(titanic_df,hue='Sex',aspect=4)
fig.map(sns.kdeplot,'Age',fill=True)
oldest = titanic_df['Age'].max()
fig.set(xlim = (0,oldest))
fig.add_legend()
```

```python
deck = titanic_df['Cabin'].dropna()
```

```python
deck.head()
```

```python
initials = []
for letter in deck:
    initials.append(letter[0])
```

```python
# Chart: Count of passengers by title (Mr, Mrs, Miss, etc.) based on name initials.
sns.catplot(x=initials,kind='count',palette='winter_d')
plt.show()
```

```python
cabin_deck =[]
for i in initials :
    if i != 'T':
        cabin_deck.append(i)
```

```python
# Chart: Count of passengers by cabin deck (inferred from cabin number).
sns.catplot(x=cabin_deck,kind='count',palette='summer')
plt.show()
```

```python
titanic_df
```

```python
# Chart: Count of embarked passengers (C = Cherbourg, Q = Queenstown, S = Southampton), split by Pclass.
sns.catplot(x='Embarked',data = titanic_df,hue='Pclass',kind='count',order=['C','Q','S'])
plt.show()
```

```python
titanic_df['Alone'] = titanic_df['SibSp'] + titanic_df['Parch']
titanic_df['Alone'] = np.where(titanic_df['Alone'] > 0, 'With Family', 'Alone')
```

```python
titanic_df
```

```python
# Chart: Shows how many passengers were traveling alone vs. with family.
sns.catplot(x='Alone',data=titanic_df,palette = 'Blues',kind='count')
```

```python
plt.show()
```

```python
titanic_df
```

```python
titanic_df['survivor'] = titanic_df['Survived'].map({0: 'no',1:'yes'})
```

```python
# Chart: Count of survivors vs non-survivors.
sns.catplot(x='survivor',data=titanic_df,palette='Set1',kind='count')
plt.show()
```

```python
# Chart: Point plot showing survival rate by Pclass and Person.
sns.catplot(x='Pclass', y='Survived', hue='Person',data=titanic_df, kind='point')
plt.show()
```

```python
# Chart: Linear regression showing relationship between age and survival.
sns.lmplot(x='Age',y='Survived',data=titanic_df)
plt.show()
```

```python
# Chart: Linear regression showing relationship between age and survival, split by Pclass.
sns.lmplot(x='Age',y='Survived',hue='Pclass',data=titanic_df)
plt.show()
'''Class 1 (●): Highest survival at all ages, gentle decline

Class 2 (●): Moderate survival, sharper decline after age 40

Class 3 (●): Lowest survival, negative survival at older ages'''
```

```python
# Chart: Binned linear regression plot of age vs survival by Pclass using predefined age groups.
gen = [10,20,40,60,80]
sns.lmplot(x='Age',y='Survived',hue='Pclass',data=titanic_df,x_bins=gen)
plt.show()
```

```python
# Chart: Binned linear regression plot of age vs survival by sex using predefined age groups.
sns.lmplot(x='Age',y='Survived',hue='Sex',data=titanic_df,x_bins=gen)
plt.show()
```
